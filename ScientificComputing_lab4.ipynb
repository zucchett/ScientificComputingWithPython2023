2. **Outer product**

Find the outer product of the following vectors:

u = np.array([1, 3, 5, 7])

v = np.array([2, 4, 6, 8])

Use different methods to do this:

* Using the function `outer` in numpy
* Using a nested `for` loop or a list comprehension
* Using numpy broadcasting operations

```python
u = np.array([1, 3, 5, 7])
v = np.array([2, 4, 6, 8])

#Using the function outer in numpy
s1 = np.outer(u,v)

print('Outer product with outer function:\n', s1)
#Using a nested for loop or a list comprehension
def for_loop(u, v):
    u_dim = u.shape[0]
    v_dim = v.shape[0]
    result = np.empty((u_dim, v_dim))
    for i in range(u_dim):
        for j in range(v_dim):
            result[i, j] = v[i]*u[j]
    return result

print('')
print('Outer product with for loop:\n', for_loop(u, v))
print('')
#Using numpy broadcasting operations
s2 = u[:, np.newaxis] * v
print('Outer product with broadcasting operations:\n', s2) 
```

7. **Prime numbers sieve**

Compute the prime numbers in the 0-N (start with N=99) range with a sieve (mask).

* Construct a shape (N,) boolean array, which is the mask
* Identify the multiples of each number starting from 2 and set accordingly the corresponding mask element
* Apply the mask to obtain an array of ordered prime numbers
* Check the performances (with `timeit`); how does it scale with N?
* Implement the optimization suggested in the sieve of Eratosthenes

```python
start = timeit.default_timer()
def find_primes(N):
    array_all = np.arange(0,N)
    def is_prime(n):
        i=2
        j= True
        while i<n:
            if n%i==0:
                j= False
                break
            else:
                i+=1
        return j

    mask = [is_prime(i) for i in range(N)]
    mask[0]=mask[1]= False
    return array_all[mask]
print('All primes under 99:', find_primes(99))
print("The difference of time with N = 99 is :", timeit.default_timer() - start)

start1 = timeit.default_timer()
find_primes(999)
print("The difference of time with N = 999 is :", timeit.default_timer() - start)

start1 = timeit.default_timer()
find_primes(9999)
print("The difference of time with N = 9999 is :", timeit.default_timer() - start)

def sieve_of_eratosthenes(N):
    boolean_arr = np.ones(N, dtype=bool)
    boolean_arr[0] = False
    boolean_arr[1] = False
    for i in range(2,int(math.sqrt(N))):
        j = i*i
        while j < N:
            boolean_arr[j] = False 
            j = j + i
    return boolean_arr

print('Sieve of Eratosthenes:',array_all[sieve_of_eratosthenes(99)])
```

8. **Diffusion using random walk**

Consider a simple random walk process: at each step in time, a walker jumps right or left (+1 or -1) with equal probability. The goal is to find the typical distance from the origin of many random walkers after a given amount of time.

Hint: create a 2D array where each row represents a walker, and each column represents a time step.

* Take 1000 walkers and let them walk for 200 steps
* Use `randint` to create a 2D array of size $walkers$ x $steps$ with values -1 or 1
* Calculate the walking distances for each walker (e.g. by summing the elements in each row)
* Take the square of the previously-obtained array (element-wise)
* Compute the mean of the squared distances at each step (i.e. the mean along the columns)
* **Optional**: plot the average distances (2) as a function of time (step)

```python
walkers = 1000
steps = 200

array_2D = np.random.randint(0, 2, size=(walkers, steps))
array_2D = 2*array_2D - 1
print('1000 walkers x 200 steps:\n', array_2D)

dist = np.cumsum(array_2D,axis=1)
print('distances:\n', dist)

sq_dist = dist**2
print('square of distances:\n', sq_dist)

sq_dist_mean = np.mean(sq_dist, axis=0)
print('mean of the squared distances:\n', sq_dist_mean)

av_dist = np.sqrt(sq_dist_mean)

plt.plot(av_dist, label='Average Distance')
plt.title('Average distances as a  function of time')
plt.legend()
plt.show()
```
