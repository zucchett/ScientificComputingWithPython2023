```python
import math
def prime_num(N):
    mask = np.ones(N+1, dtype=bool)
    mask[0] = False
    mask[1] = False
    for i in range(2,int(math.sqrt(N))+1):
        if mask[i]:
            for j in range(i**2, N+1, i):
                mask[j] = False
    
    num = np.arange(N+1)
    primes = num[mask]
    return primes
            
print(prime_num(1000), '\n')
#----second part
%timeit prime_num(100)
%timeit prime_num(1000)
%timeit prime_num(10000)
print('\nas N grows,the time consumption to calculate prime numbers extremely grow.')
print('it is expectable, due to iterating.')
```
