3\. **Profile plot**

Produce a profile plot from a scatter plot.
* Download the following pickle file:
```bash
wget https://www.dropbox.com/s/3uqleyc3wyz52tr/residuals_261.pkl -P data/
```
* Inspect the dataset, you'll find two variables (features)
* Convert the content to a Pandas Dataframe
* Clean the sample by selecting the entries (rows) with the absolute values of the variable "residual" smaller than 2
* Plot a Seaborn `jointplot` of "residuals" versus "distances", and use seaborn to display a linear regression. 

Comment on the correlation between these variables.

* Create manually (without using seaborn) the profile histogram for the "distance" variable; choose an appropriate binning.
* Obtain 3 numpy arrays:
  * `x`, the array of bin centers of the profile histogram of the "distance" variable
  * `y`, the mean values of the "residuals", estimated in slices (bins) of "distance"
  * `err_y`, the standard deviation of the of the "residuals", estimated in slices (bins) of "distance"
* Plot the profile plot on top of the scatter plot

````python
# Download the dataset
#!wget https://www.dropbox.com/s/3uqleyc3wyz52tr/residuals_261.pkl -P ./data/

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset from the pickle file
dataset = np.load('data/residuals_261.pkl', allow_pickle=True).item()
dataset = pd.DataFrame(dataset)

# Filter the dataset by selecting rows with residuals < 2
dataset = dataset[abs(dataset['residuals']) < 2]

# Calculate and print the correlation between residuals and distances
corr = dataset[['residuals', 'distances']].corr()
print("Correlation between residuals and distances:\n", corr)

# Create a jointplot with regression line
joint_plot = sns.jointplot(x='distances', y='residuals', data=dataset, kind="reg", joint_kws={'line_kws': {'color': 'b', 'label': 'Regression'}})
joint_plot.ax_joint.legend()
joint_plot.fig.set_size_inches(8, 8)
plt.title("Jointplot with Regression Line")
plt.show()

# Plot the histogram of distances
bin_count = 25
histogram = plt.hist(dataset['distances'], bins=bin_count, alpha=0.5, edgecolor='white', range=(0, 20))
plt.title("Histogram of Distances")
plt.show()

# Calculate the profile plot data
bin_centers = (histogram[1][1:] + histogram[1][:-1]) * 0.5
print(f"Bin Centers:\n{bin_centers}")

mean_values = [dataset["residuals"][(dataset['distances'] >= histogram[1][i]) & (dataset['distances'] <= histogram[1][i + 1])].mean() for i in range(bin_count)]
print(f"\nMean Values:\n{mean_values}")

std_dev_values = [dataset["residuals"][(dataset['distances'] > histogram[1][i]) & (dataset['distances'] < histogram[1][i + 1])].std() for i in range(bin_count)]
print(f"\nStandard Deviation Values:\n{std_dev_values}")

# Use the same jointplot object for the profile plot
joint_plot.fig.set_size_inches(10, 10)
plt.sca(joint_plot.ax_joint)
plt.errorbar(x=bin_centers, y=mean_values, yerr=std_dev_values, marker='o', color='r', xlolims=22, label="Profile")
plt.legend()
plt.title("Profile Plot on Top of Scatter Plot")
plt.show()
````

4\. **Kernel Density Estimate**

Produce a KDE for a given distribution (by hand, not using seaborn):

* Fill a numpy array `x` of length N (with $N=\mathcal{O}(100)$) with a variable normally distributed, with a given mean and standard deviation
* Fill an histogram in pyplot taking proper care of the aesthetic:
   * use a meaningful number of bins
   * set a proper y axis label
   * set proper value of y axis major ticks labels (e.g. you want to display only integer labels)
   * display the histograms as data points with errors (the error being the poisson uncertainty)
* For every element of `x`, create a gaussian with the mean corresponding to the element value and the standard deviation as a parameter that can be tuned. The standard deviation default value should be:
$$ 1.06 * x.std() * x.size ^{-\frac{1}{5}} $$
you can use the scipy function `stats.norm()` for that.
* In a separate plot (to be placed beside the original histogram), plot all the gaussian functions so obtained
* Sum (with `np.sum()`) all the gaussian functions and normalize the result such that the integral matches the integral of the original histogram. For that you could use the `scipy.integrate.trapz()` method. Superimpose the normalized sum of all gaussians to the first histogram.


````python
from scipy import stats
import scipy
from matplotlib.ticker import AutoMinorLocator

mean_val = 10
std_dev = 0.5
data_size = 100
num_bins = int(data_size / 4)
data_values = mean_val + std_dev * np.random.randn(data_size)

# Create subplots
fig, (histogram_ax, gaussians_ax) = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Plot Histogram with Error Bars
hist, bin_edges, _ = histogram_ax.hist(data_values, bins=num_bins, color='magenta', edgecolor='black')
histogram_ax.set_title('Histogram')
histogram_ax.set_xlabel('Values')
histogram_ax.set_ylabel('Occurrences')
histogram_ax.set_yticks(np.arange(0, data_size, 1))

error = np.sqrt(hist)
bins_center = (bin_edges[:-1] + bin_edges[1:]) * 0.5
histogram_ax.errorbar(bins_center, hist, yerr=error, fmt='none', color='red', label="Error Bars")

# Gaussian Functions for each element of x
std_def = 1.06 * data_values.std() * data_values.size**(-1/5)
gaussians = []
space = np.linspace(np.min(data_values), np.max(data_values), data_size)
for item in data_values:
    normal_distribution = stats.norm(item, std_def)
    gaussians.append(normal_distribution.pdf(space))
    gaussians_ax.plot(space, gaussians[-1])

gaussians_ax.set_title("Gaussian for every element of x")
gaussians_ax.set_xlabel('Values')
gaussians_ax.set_ylabel('Probability Density')

# Kernel Density Estimation (KDE)
area_sog = scipy.integrate.trapz(np.sum(gaussians, axis=0), space)
area_hist = sum(np.diff(bin_edges) * hist)
k_norm = area_sog / area_hist
norm_sog = np.sum(gaussians, axis=0) / k_norm

# Plot KDE
histogram_ax.set_title('Kernel Density Estimation')
histogram_ax.set_xlabel('Values')
histogram_ax.plot(space, norm_sog, color="green", label="KDE")
histogram_ax.legend()

plt.show()
````
